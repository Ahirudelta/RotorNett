```python
import numpy as np
import math
```

```python

N = 16                                         # network endpoints
Nreal = 93                                     # number of demand matrix realizations of each # flows
Ncycles = 4

## Set up matchings
# assign number of switches (NS) and number of matchings per switch
# (Nmatch):

Nmatch = 4                                     # number of matchings per switch (crossover matchings)
NS = math.ceil(N/Nmatch)                       # number of switches

# construct bidirectional matchings

mc = [0]*int(np.log2(N))
for a in range(1,int(np.log2(N)) + 1):
    mc[a-1] = np.zeros(shape = (N,N))
    for b in range(1,int(N/(2**(a+1))) + 1):
        for c in range(0,2**a):
            mc[a-1][((2**a)*b-c) - 1][((2**a)*b-((2**a-1)-c)) - 1] = 1

mc = np.array(mc)
C = [0]*3
C[0] = [1,2,3]
C[1] = [1,2,3,4]
C[2] = [4]
NC = len(C)

m = [0]*Nmatch                                 # all connections seen at each of Nmatch timeslots
for a in range(0,Nmatch):
    m[a] = mc[a].copy()
    for b in range(0,NC):
        n = len(C[b])
        temp = np.eye(N)
        for c in range(0,n):
            temp = temp*mc[C[b][c] - 1]
        m[a] = m[a] + temp*mc[a]
    
    # Ensure main diaonal is zero
    for i in range(0,len(m[a])):
        for j in range(0,len(m[a][0])):
            if i == j:
                m[a][i][j] = 0

## Main loop

Nflow0 = [i for i in range(1,N**2-N + 1,10)]   # setup to sweep number of non-zero flows
if Nflow0[-1] != N**2-N:
    Nflow0.append(N**2-N)

NN = len(Nflow0)

Bwr = np.zeros(shape = (NN, Nreal, Ncycles))

for y in range(0, Nreal):
    for x in range(0, NN):
        print("Iteration: y = %d, x = %d" %(y,x))
        
        #generate a random demand matrix:
        r = [0,0]
        r[0] = np.ones(shape = (1,Nflow0[x]))
        r[1] = np.zeros(shape = (1,N**2-N-Nflow0[x]))
        print(r)
        #r = np.array(r)
        #l = [len(r[0]), len(r[0][0]), len(r[1]), len(r[1][0])]
        #l = max(l)
        #print(l)
        #r = r(np.random.permutation(range(l)))
```
